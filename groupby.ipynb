#### groupby and dropna

```python
#by default a group is droped if its value is null. by using dropna=False, we can retain a group
data = [[10, 20, 30], [100, None, 400], [25, 15, 35], [11, 22, 22]]

df = pd.DataFrame(data, columns=["a", "b", "c"])
df
```

```python
df.groupby('b').sum()
```

```python
df.groupby('b',dropna=False).sum()
```

```python
df.groupby('b',dropna=False,sort=False).sum()
```
