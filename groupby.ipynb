#### groupby and dropna

```python
#by default a group is droped if its value is null. by using dropna=False, we can retain a group
data = [[10, 20, 30], [100, None, 400], [25, 15, 35], [11, 22, 22]]

df = pd.DataFrame(data, columns=["a", "b", "c"])
df
```

```python
df.groupby('b').sum()
```

```python
df.groupby('b',dropna=False).sum()
```

```python
df.groupby('b',dropna=False,sort=False).sum()
```

```python
g=tips.groupby(['sex','day'])
```

```python
g['tip'].agg([np.sum,np.min,np.max]).rename(columns={'sum':'total_tip','amin':'tip_min','amax':'tip_max'})
```

```python
g[['tip','total_bill']].agg([np.sum,np.min,np.max])
```

```python
g[['tip','total_bill']].agg([np.sum,np.min,np.max]).rename(columns={'sum':'total','amin':'min','amax':'max'})
```

```python
g[['tip','total_bill']].agg([lambda x:x.max()-x.min(),lambda x:x.median()-x.mean()])
```

#### Named Aggregation  --->   groupby.agg()

```python
tips.groupby(['sex','day'])[['tip','total_bill']].agg(max_tip=pd.NamedAgg(column="tip",aggfunc="max"),
                                                    max_bill=pd.NamedAgg(column="total_bill",aggfunc="max")).reset_index()
```

```python
tips.groupby(['sex','day'])[['tip','total_bill']].agg(max_tip=("tip","max"),
                                                    max_bill=("total_bill","max")).reset_index()
```

```python
tips.groupby(['sex','day'])[['tip','total_bill']].agg(max_tip=("tip","max"),
                                                      max_bill=("total_bill","max"),
                                                      min_bill=("total_bill","min"),
                                                    max_min_diff=("total_bill",lambda x:x.max()-x.min())).reset_index()
```

```python
tips.groupby(['sex','day'])[['total_bill','tip']].agg({'total_bill':[sum,max],'tip':max})
```

```python
T1=tips.groupby(['sex','day'])[['total_bill','tip']].agg({'total_bill':[sum,max],'tip':max}).reset_index()
```
